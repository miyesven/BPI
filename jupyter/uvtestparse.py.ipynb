```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('UV/Aug212021_GoodTestsMatrix.csv')
df = df.transpose()
df.head()
```

```python
lbd = [ float(i) for i in df.index.values.tolist()[1:] ]
foundation = df.iloc[:,0].tolist()[1:]
microscopeslide = df.iloc[:,1].tolist()[1:]
ziploc = df.iloc[:,2].tolist()[1:]
qzglass = df.iloc[:,3].tolist()[1:]
qzg_ALcoating = df.iloc[:,4].tolist()[1:]
qzg_109030ALLC = df.iloc[:,5].tolist()[1:]
ziploc_208030LC = df.iloc[:,6].tolist()[1:]
ziploc_109030 = df.iloc[:,7].tolist()[1:]
qzg_109030_airbrush = df.iloc[:,8].tolist()[1:]
qzg_30pcLS = df.iloc[:,9].tolist()[1:]
qzg_109030ALLC_coatlightly = df.iloc[:,10].tolist()[1:]
qzg_109030AL_coatlightly = df.iloc[:,11].tolist()[1:]
qzg_109030LS_coatlightly = df.iloc[:,12].tolist()[1:]
```

```python
np.arange(200,700,50)
```

```python
fig = plt.figure(figsize=(15,10))
plt.rcParams.update({'font.size': 12})

ax1 = plt.subplot(221)
# ax1.plot(lbd, foundation, label='Ayako Foundation')
ax1.plot(lbd, microscopeslide, label='Microscope Slide', color='orange')
ax1.plot(lbd, ziploc, label='Ziploc Bag', color='darkgreen')
ax1.plot(lbd, qzglass, label='Quartz Glass', color='steelblue')
ax1.legend()
ax1.set_xlim(200,500)
ax1.set_xlabel('Wavelength (nm)')
ax1.set_ylabel('Transmittance (%)')

ax2 = plt.subplot(222)
ax2.plot(lbd, qzg_ALcoating, label='AL coating', color='sienna')
ax2.plot(lbd, qzg_30pcLS, label='LS coating', color='chocolate')
ax2.plot(lbd, qzg_109030_airbrush, label='90% LS Airbrushed', color='goldenrod')
ax2.legend()
ax2.set_xlim(200,500)
ax2.set_xlabel('Wavelength (nm)')
ax2.set_ylabel('Transmittance (%)')

ax3 = plt.subplot(212)
ax3.plot(lbd, qzglass, label='Quartz Glass', color='steelblue')
ax3.plot(lbd, qzg_109030AL_coatlightly, label='90% AL coat', color='darkgoldenrod')
ax3.plot(lbd, qzg_109030LS_coatlightly, label='90% LS coat', color='goldenrod')
ax3.plot(lbd, foundation, label='Ayako Foundation', color='lightsalmon')
ax3.legend()
ax3.set_xlim([200,500])
ax3.set_xlabel('Wavelength (nm)')
ax3.set_ylabel('Transmittance (%)')
ax3.vlines(320,-5,90, linestyles='dashed', color='plum')
ax3.vlines(380,-5,90, linestyles='dashed', color='plum')
ax3.vlines(280,-5,90, linestyles='dashed', color='plum')

# plt.minor_ticks_off()
plt.title('UV Transmittance of Lignin/PLA W/O Emulsion')

plt.savefig('UVTests_Aug212021.png')
```
