```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Final Test Results, coatings and UV test done by Ayako. These were done with the 10mm quartz glass and the 1mm quartz glass in the background and sample holder in the UV Vis machine respectively. The intended sample is diluted with chloroform and brushed onto one side of the quartz glass curvette. Despite the difference in thickness, the 1mm was verified to have 100% transparency against the background of the 10mm

```python
columns = ['Wavelength', 'lg90_rub1', 'lg90_brush1', \
           'lg60_brush1', 'lg60_brush2', 'lg60_brush3', 'lg60_brush4',\
           'lg90_rub2',\
           'qzg',\
           'lg90_brush2', 'lg90_brush3', 'lg90_brush4', 'lg90_brush5']
df = pd.read_csv('UV/goodtestsfinal.txt', delimiter='\t+', names=columns, skiprows=1)
df.head()
```

```python
plt.figure(figsize=(20,15))
for column in df:
    if column != 'Wavelength':
        plt.plot(df['Wavelength'], df[column]/df[column].iloc[-1], label=column)
plt.legend()
plt.savefig("UV/uvoversall.png")
```

Final Data Used
For 60% lignin sample, I used "Ayako_406025ALdiluted_brushed_0.0003g_23Aug2021 - RawData"
And for 90% lignin sample, "Ayako_10907AF_Diluted_blushed_0.0005g(2layers)_1Sep2021 - RawData"

These are 'lg60_brush4' and 'lg90_brush4' respectively. 

```python
plt.figure(figsize=(15,10))
plt.title('UV Test of PLA-Lignin Emulsion on Quartz Glass')
plt.xlabel('Wavelength (nm)', fontsize=20)
plt.ylabel('Normalized Transmittance', fontsize=20)
# UVA 315-400, UVB 280-315, UVC 100-280
from matplotlib.patches import Rectangle
UVA_min, UVA_max = 315, 400
UVB_min, UVB_max = 280, 315
UVC_min, UVC_max = 200, 280
currentAxis = plt.gca()
currentAxis.add_patch(Rectangle((UVA_min, 0), UVA_max-UVA_min, 1, facecolor='mediumpurple', alpha=0.1))
currentAxis.add_patch(Rectangle((UVB_min, 0), UVB_max-UVB_min, 1, facecolor='blueviolet', alpha=0.15))
currentAxis.add_patch(Rectangle((UVC_min, 0), UVC_max-UVC_min, 1, facecolor='rebeccapurple', alpha=0.2))
plt.text(340, 0.9, 'UV-A', fontsize = 22)
plt.text(285, 0.9, 'UV-B', fontsize = 22)
plt.text(230, 0.9, 'UV-C', fontsize = 22)

plt.plot(df['Wavelength'], df['lg60_brush4']/df['lg60_brush4'].iloc[-1], label='60-30-7', marker='o', markersize=3, linewidth=0)
plt.plot(df['Wavelength'], df['lg90_brush4']/df['lg90_brush4'].iloc[-1], label='90-35-7', marker='o', markersize=3, linewidth=0)
plt.plot(df['Wavelength'], df['qzg']/df['qzg'].iloc[-1], label='Quartz Glass', marker='o', markersize=3, linewidth=0)
plt.legend()
plt.rcParams.update({'font.size': 16})
plt.savefig('UV/UV_graph.png')
```

```python
df = pd.read_csv('UV/Aug212021_GoodTestsMatrix.csv')
df = df.transpose()
df.head()
```

```python
lbd = [ float(i) for i in df.index.values.tolist()[1:] ]
foundation = df.iloc[:,0].tolist()[1:]
microscopeslide = df.iloc[:,1].tolist()[1:]
ziploc = df.iloc[:,2].tolist()[1:]
qzglass = df.iloc[:,3].tolist()[1:]
qzg_ALcoating = df.iloc[:,4].tolist()[1:]
qzg_109030ALLC = df.iloc[:,5].tolist()[1:]
ziploc_208030LC = df.iloc[:,6].tolist()[1:]
ziploc_109030 = df.iloc[:,7].tolist()[1:]
qzg_109030_airbrush = df.iloc[:,8].tolist()[1:]
qzg_30pcLS = df.iloc[:,9].tolist()[1:]
qzg_109030ALLC_coatlightly = df.iloc[:,10].tolist()[1:]
qzg_109030AL_coatlightly = df.iloc[:,11].tolist()[1:]
qzg_109030LS_coatlightly = df.iloc[:,12].tolist()[1:]
```

```python
np.arange(200,700,50)
```

```python
fig = plt.figure(figsize=(15,10))
plt.rcParams.update({'font.size': 12})

ax1 = plt.subplot(221)
# ax1.plot(lbd, foundation, label='Ayako Foundation')
ax1.plot(lbd, microscopeslide, label='Microscope Slide', color='orange')
ax1.plot(lbd, ziploc, label='Ziploc Bag', color='darkgreen')
ax1.plot(lbd, qzglass, label='Quartz Glass', color='steelblue')
ax1.legend()
ax1.set_xlim(200,500)
ax1.set_xlabel('Wavelength (nm)')
ax1.set_ylabel('Transmittance (%)')

ax2 = plt.subplot(222)
ax2.plot(lbd, qzg_ALcoating, label='AL coating', color='sienna')
ax2.plot(lbd, qzg_30pcLS, label='LS coating', color='chocolate')
ax2.plot(lbd, qzg_109030_airbrush, label='90% LS Airbrushed', color='goldenrod')
ax2.legend()
ax2.set_xlim(200,500)
ax2.set_xlabel('Wavelength (nm)')
ax2.set_ylabel('Transmittance (%)')

ax3 = plt.subplot(212)
ax3.plot(lbd, qzglass, label='Quartz Glass', color='steelblue')
ax3.plot(lbd, qzg_109030AL_coatlightly, label='90% AL coat', color='darkgoldenrod')
ax3.plot(lbd, qzg_109030LS_coatlightly, label='90% LS coat', color='goldenrod')
ax3.plot(lbd, foundation, label='Ayako Foundation', color='lightsalmon')
ax3.legend()
ax3.set_xlim([200,500])
ax3.set_xlabel('Wavelength (nm)')
ax3.set_ylabel('Transmittance (%)')
ax3.vlines(320,-5,90, linestyles='dashed', color='plum')
ax3.vlines(380,-5,90, linestyles='dashed', color='plum')
ax3.vlines(280,-5,90, linestyles='dashed', color='plum')

# plt.minor_ticks_off()
plt.title('UV Transmittance of Lignin/PLA W/O Emulsion')

plt.savefig('UVTests_Aug212021.png')
```
